```python
members.drop(['registration_init_time'],axis=1,inplace=True)
user_logs.drop(['date'],axis=1,inplace=True)
transactions.drop(['transaction_date'],axis=1,inplace=True)
transactions.drop(['membership_expire_date'],axis=1,inplace=True)
```

```python
#transaction features
```

```python
#discount
transactions['discount'] = transactions['plan_list_price'] - transactions['actual_amount_paid']
#is_discount
transactions['is_discount'] = transactions.discount.apply(lambda x: 1 if x > 0 else 0)
#amount per day 
transactions['amt_per_day'] = transactions['actual_amount_paid'] / transactions['payment_plan_days']
#membership_duration
transactions['membership_duration'] = transactions.membership_expire_date_fixed - transactions.transaction_date_fixed
transactions['membership_duration'] = transactions['membership_duration'] / np.timedelta64(1, 'D')
transactions['membership_duration'] = transactions['membership_duration'].astype(int)
```

```python
members_v01 = members[members['gender'].notnull()]
members_v01_sample = members_v01.sample(frac=0.009)
```

```python
members_v01_sample.shape
```

```python
members_v01_sample.head()
```

```python
%%time
transactions_subset = transactions[transactions['msno'].isin(list(members_v01_sample['msno']))]
user_logs_subset = user_logs[user_logs['msno'].isin(list(members_v01_sample['msno']))]
```

```python
%%time
def get_features(data):
#     print(data['msno'])
    membership_expire_date_last = transactions_subset[transactions_subset['msno']==data['msno']]['membership_expire_date_fixed'].max()
    total_transactions = transactions_subset[transactions_subset['msno']==data['msno']].shape[0]
    total_activaity = user_logs_subset[user_logs_subset['msno']==data['msno']].shape[0]
    return membership_expire_date_last, total_transactions, total_activaity
    

members_v01_sample[['membership_expire_date_last','total_transactions','total_activity']] = members_v01_sample.apply(get_features,axis=1, result_type="expand")
```

```python
members_v01_sample.head()
```

```python
%%time
members_v01_sample_transform_01 = members_v01_sample[members_v01_sample['membership_expire_date_last'].notnull()]
members_v01_sample_transform_01['registration_init_time_fixed'] = pd.to_datetime(members_v01_sample_transform_01['registration_init_time_fixed'])
members_v01_sample_transform_01['membership_expire_date_last'] = pd.to_datetime(members_v01_sample_transform_01['membership_expire_date_last'])
members_v01_sample_transform_01['time_on_platform_months'] = members_v01_sample_transform_01['membership_expire_date_last'].dt.to_period('M').astype(int)-members_v01_sample_transform_01['registration_init_time_fixed'].dt.to_period('M').astype(int)
members_v01_sample_transform_01.drop(['registration_init_time_fixed','membership_expire_date_last'], axis=1, inplace=True)
```

```python
%%time
# Joins  -> train  -> user_logs -> transactions
members_v01_sample_join01 = members_v01_sample_transform_01.merge(train,on='msno')
user_logs_groupby_subset = user_logs[user_logs['msno'].isin(list(members_v01_sample_join01['msno']))].groupby(by='msno').mean()
members_v01_sample_join02 = members_v01_sample_join01.merge(user_logs_groupby_subset,on='msno')
transactions_groupby_subset = transactions[transactions['msno'].isin(list(members_v01_sample_join01['msno']))].groupby(by='msno').mean()
members_v01_sample_joinfinal = members_v01_sample_join02.merge(transactions_groupby_subset,on='msno')
```
